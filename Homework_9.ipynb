### Step 0

Run the necessary imports.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
import dlib
```

### Step 1

Load an image (any image that contains faces).

```python
img = cv2.imread('friends.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img)
```

### Step 2

Load the dlib face predictor.

```python
detector = dlib.get_frontal_face_detector()
```

### Step 3

Run the predictor on your image.

```python
rects = detector(gray, 1)

print('Number of detected faces:', len(rects))

def rect_to_bb(rect):
    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y

    return (x, y, w, h)
```

### Step 4

Draw bounding boxes around the detected faces and plot the image. Use different colour for each face.

```python
# Making a copy of the original image to draw rectangles
result_dlib = np.copy(img)

# Defining colors for rectangles
colors = [(255, 0, 0), (0, 255, 255), (0, 255, 0), (0, 0, 255), (255, 255, 0), (255, 140, 0),(255, 0, 255), (255, 255, 255), (255, 105, 180), (127, 255, 212)]
faces_dlib_img = []

for i, rect in enumerate(rects):
    x, y, w, h = rect_to_bb(rect)
    color = colors[i % len(colors)]
    cv2.rectangle(result_dlib, (x, y), (x+w, y+h), color, 3)
    faces_dlib_img.append(img[y:y+h, x:x+w, :])

# Displaying the result
plt.subplot(122), plt.imshow(result_dlib)
plt.show()
```
